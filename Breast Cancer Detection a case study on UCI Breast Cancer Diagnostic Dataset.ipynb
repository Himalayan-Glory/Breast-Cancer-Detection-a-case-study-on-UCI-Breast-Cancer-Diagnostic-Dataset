#### Model Training

```python
X=df_cancer.drop(['target'],axis=1)
```

```python
X
```

```python
y=df_cancer['target']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=5)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
svc_model=SVC()
```

```python
svc_model.fit(X_train,y_train)
```

#### Evaluating the model

```python
y_predict=svc_model.predict(X_test)
```

```python
y_predict
```

```python
cm=confusion_matrix(y_test,y_predict)
```

```python
cm
```

```python
sns.heatmap(cm)
```

```python
sns.heatmap(cm,annot=True)
```

```python
sns.heatmap(cm,annot=True,cmap=plt.get_cmap('gray'))
```

```python
print(classification_report(y_test,y_predict))
```

#### Improving the model performance

```python
#normalizing i.e changing values range 0-1
```

```python
min_train=X_train.min()
```

```python
min_train
```

```python
max_train=X_train.max()
```

```python
max_train
```

```python
#range_train=(X_train-min_train).max()
```

```python
#range_train
```

```python
range_train=max_train-min_train
```

```python
range_train
```

```python
X_train_normalized=(X_train-min_train)/range_train
```

```python
X_train_normalized
```

```python
sns.scatterplot(x=X_train['mean area'],y=X_train['mean smoothness'],hue=y_train)
```

```python
sns.scatterplot(x=X_train_normalized['mean area'],y=X_train_normalized['mean smoothness'],hue=y_train)
```

```python
#normalizing X_test 
min_test=X_test.min()
max_test=X_test.max()
X_test_normalized=(X_test-min_test)/(max_test-min_test)
```

```python
svc_model.fit(X_train_normalized,y_train)
```

```python
y_predict_new=svc_model.predict(X_test_normalized)
```

```python
cm=confusion_matrix(y_test,y_predict_new)
```

```python
cm
```

```python
sns.heatmap(cm,annot=True,cmap=plt.get_cmap('pink'))
```

```python
print(classification_report(y_test,y_predict_new))
```

#### further improvement 

```python
#grid search for best hyper parameters (c,gamma,kernel)
```

```python
param_grid={'C':[0.1,1,10,100],'gamma':[1,0.1,0.01,0.001],'kernel':['rbf']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid=GridSearchCV(SVC(),param_grid,refit=True,verbose=4)
```

```python
grid.fit(X_train_normalized,y_train)
```

```python
grid.best_params_
```

```python
prediction_with_gridsearch=grid.predict(X_test_normalized)
```

```python
cm_new=confusion_matrix(y_test,prediction_with_gridsearch)
```

```python
cm_new
```

```python
sns.heatmap(cm_new,annot=True)
```

```python
print(classification_report(y_test,prediction_with_gridsearch))
```
